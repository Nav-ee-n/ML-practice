Machine Learning


```python
# machine learning

# ----Types of learning--------------
# labeled data - data which are already defined
# class label / target / dependent variable / prediction / output - the output which defines the result

# 1.supervised learning - learning using labeled data / learning under the supervision of someone

# 2.unsupervised learning - no class label

# 3.reinforcement learning - reward & punishment

# machine learning is different from normal coding as in ML we give the input and output. After which we give a new input to predict the o/p
# features / attributes / independent variable / input / predictors - are columns which are apart from the class label
# data - rows / records / objects / data point
```
